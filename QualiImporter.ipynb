```python
links = ['https://en.wikipedia.org/wiki/2022_Firestone_Grand_Prix_of_St._Petersburg',
    'https://en.wikipedia.org/wiki/2022_XPEL_375',
    'https://en.wikipedia.org/wiki/2022_Acura_Grand_Prix_of_Long_Beach',
    'https://en.wikipedia.org/wiki/2022_Honda_Indy_Grand_Prix_of_Alabama',
    'https://en.wikipedia.org/wiki/2022_GMR_Grand_Prix',
    'https://en.wikipedia.org/wiki/2022_Indianapolis_500',
    'https://en.wikipedia.org/wiki/2022_Chevrolet_Detroit_Grand_Prix',
    'https://en.wikipedia.org/wiki/2022_Sonsio_Grand_Prix_at_Road_America',
    'https://en.wikipedia.org/wiki/2022_Honda_Indy_200',
    'https://en.wikipedia.org/wiki/2022_Honda_Indy_Toronto',
    'https://en.wikipedia.org/wiki/2022_Hy-Vee_IndyCar_Race_Weekend',
    'https://en.wikipedia.org/wiki/2022_Gallagher_Grand_Prix',
    'https://en.wikipedia.org/wiki/2022_Big_Machine_Music_City_Grand_Prix',
    'https://en.wikipedia.org/wiki/2022_Bommarito_Automotive_Group_500',
    'https://en.wikipedia.org/wiki/2022_Grand_Prix_of_Portland',
    'https://en.wikipedia.org/wiki/2022_Firestone_Grand_Prix_of_Monterey']
df_quali = df[['Driver', 'Rookie', 'Nr', 'Team']].copy()
j=0
for i, link in enumerate(links):
    if i==0 or 1 < i < 5 or 5 < i < 10 or 11 < i < 13 or 13 < i:
        tables = pd.read_html(link)
        quali_results = tables[7][['Driver', 'Grid']].drop(tables[7].tail(2).index, axis=0)
    elif i==1 or i==11 or i==13:
        tables = pd.read_html(link)
        quali_results = tables[6][['Driver', 'Grid']].drop(tables[6].tail(2).index, axis=0)
    elif i==5:
        tables = pd.read_html(link)
        quali_results = tables[22][['Driver', 'Grid']].drop(tables[22].tail(2).index, axis=0)
    elif i==10:
        tables = pd.read_html(link)
        quali_results = tables[5][['Driver', 'Grid']].drop(tables[5].tail(2).index, axis=0)
        
    quali_results['Driver'] = quali_results['Driver'].str.replace('\s+R$', '')
    quali_results['Driver'] = quali_results['Driver'].str.replace('\s+W$', '')
    quali_results['Driver'] = quali_results['Driver'].str.replace(' *\[[a-z]* *[0-9]*\]', '')
    df_quali = df_quali.merge(quali_results, on='Driver', how='outer')
    quali_results = quali_results.rename(columns={'Grid':columns[i+j+4]})
    # df_quali = df_quali.merge(quali_results, on='Driver', how='outer')
    df_quali = pd.merge(df_quali, quali_results, how='outer', on='Driver')
    if i == 10:
        quali_results = tables[12][['Driver', 'Grid']].drop(tables[12].tail(2).index, axis=0)
        j+=1
        quali_results['Driver'] = quali_results['Driver'].str.replace('\s+R$', '')
        quali_results['Driver'] = quali_results['Driver'].str.replace('\s+W$', '')
        quali_results = quali_results.rename(columns={'Grid':columns[i+j+4]})
        df_quali = pd.merge(df_quali, quali_results, how='outer', on='Driver')
    # print(df_quali)
df_quali = df_quali.drop([i for i in df_quali.columns if 'Grid' in i], axis=1)
```
