```python
import pandas as pd
import numpy as np
```

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/2014_IndyCar_Series')
tables[1]
```

```python
season = 2014
lineups = tables[1].copy()
calendar = tables[2]
results = tables[3]
# calendar = pd.concat([calendar.iloc[:6], calendar.iloc[5:]], axis=0, ignore_index=True)
results
```

```python
# lineups['Drivers'] = lineups['Drivers'].str.replace(' \(R\)', '')
lineups = lineups.rename(columns = {'Driver(s)':'Drivers'})
lineups['Drivers'] = lineups['Drivers'].str.replace('\s+R$', '')
lineups['Drivers'] = lineups['Drivers'].str.replace('\s+R\s+', '')
lineups['Drivers'] = lineups['Drivers'].str.replace(' *\[[A-Z]* *[1-9]*\]', '')
lineups['Engine'] = lineups['Engine'].str.replace(' *\[[A-Z]* *[1-9]*\]', '')
lineups['Team'] = lineups['Team'].str.replace(' *\[[1-9]*\]', '')
lineups.at[19, 'Engine'] = 'Chevrolet'
lineups
```

```python
# results.columns = results.columns.droplevel()
# results = results.rename(columns = {'Winning driver':'Winner'})
results = results.rename(columns = {'Driver':'Winner', 'Team':'Winning Team', 'Manufacturer':'Winning Engine'})
results['Pole position'] = results['Pole position'].str.replace(' *\[[A-Z]* *[1-9]*\]', '')
# results.at[16, 'Pole position'] = 'Takuma Sato'
results
```

```python
lineups[lineups['Drivers'].isin(list(results['Winner'].unique()))]
# list(tables[4]['Winner'].unique())
# tables[2]['Drivers'].isin(list(tables[4]['Winner'].unique()))
```

```python
df = pd.DataFrame(columns=['Season', 'Race', 'Track', 'Type', 'Pole Position', 'Winner', 'Winning Team', 'Winning Engine'])
# df[['Race', 'Track']] = calendar[['Race Name', 'Track']]
df[['Race', 'Track']] = calendar[['Race name', 'Track']]
df['Type'] = "Oval"
df[['Pole Position', 'Winner']] = results[['Pole position', 'Winner']]
for index, row in df.iterrows():
    try:
        row[['Winning Team', 'Winning Engine']] = list(lineups[lineups['Drivers']==row['Winner']].iloc[0][['Team', 'Engine']])
    except KeyError:
        row['Winning Team'] = list(lineups[lineups['Drivers']==row['Winner']]['Team'])[0]
        row['Winning Engine'] = "Honda"
    except IndexError:
        pass
df['Season'] = season
df['Race'] = df['Race'].str.replace(' *\[[A-Z]*[a-z]* *[1-9]*\]', '')
df
```

```python
# df.at[0, 'Track'] = 'Streets of St. Petersburg R'
df.at[3, 'Race'] = 'Grand Prix of Indianapolis'
# df
```

```python
current_file = pd.read_csv('IRL_Indycar.csv')
new_file = pd.concat([current_file, df], axis=0)
new_file.index = range(len(new_file))
new_file.to_csv('IRL_Indycar.csv', mode='w', header=True, index=False)
new_file
```
